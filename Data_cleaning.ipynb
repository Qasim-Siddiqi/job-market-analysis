```python
pay_split[0]= pay_split[0].str.extract(r'(\d+)')[0] # .str.extract() returns a DataFrame of groups.
                                                    #[0] selects the first (and only) column from that DataFrame.
```

```python
#pay_split[1] = pay_split[1].astype(str).str.replace(r'^(\d+).*', r'\1', regex=True) # ^(\d+) captures one or more digits at the beginning of the string.
# .* matches everything after that.
# The whole match (^(\d+).*) is replaced with only the captured digits (\1), effectively removing everything else in the string
 # \1 means: “replace the whole string with the first captured group”
```

```python
df['max_exp']=df['max_exp'].str.replace('yrs', '').str.strip()
df['min_exp'] = df['min_exp'].replace('Not Mentioned', np.nan)
```

```python
df['min_exp'] = df['min_exp'].replace('', np.nan)
df['min_exp']= df['min_exp'].astype(float)
```

```python
df['max_exp']= df['max_exp'].astype(float)
```

```python
df['avg_pay']= (df['min_pay']+df['max_pay'])/2
```

```python
df['avg_exp']= (df['min_exp']+df['max_exp'])/2
```
